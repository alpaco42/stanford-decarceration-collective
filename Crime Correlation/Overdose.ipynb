```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def import_csv(csv):
    df = pd.read_csv(csv, header=1)
    df = df[df.columns[0:6]]
    df.columns = ["County" if x == "County / Year" else x for x in df.columns]
    df["County"] = df["County"].str.strip()
    # Slight hard-code hacks to fix spelling mistakes
    df["County"][37] = "San Francisco"
    df["County"][26] = "Monterey"
    df["County"][39] = "San Luis Obispo"
    return df
```

```python
def graph_district(df, name):
    district = df[df["County"] == name].iloc[0][1:]
    plt.plot(range(2018, 2023), district, label=name)
    plt.legend(bbox_to_anchor=(1, 1))
    plt.xticks(range(2018, 2022))
```

```python
def graph_bucket(df, bucket):
    for district in bucket: 
        graph_district(df, district)
```

```python
# Define all 10 buckets
bucket_1 = ["Los Angeles", "San Diego", "Orange", "Riverside", "San Bernardino", "Santa Clara"]
bucket_2 = ["Alameda", "Sacramento", "Contra Costa", "Fresno", "Kern", "San Francisco"]
bucket_3 = ["Ventura", "San Joaquin", "San Mateo", "Stanislaus", "Sonoma", "Tulare"]
bucket_4 = ["Solano", "Santa Barbara", "Monterey", "Placer", "San Luis Obispo", "Merced"]
bucket_5 = ["Santa Cruz", "Marin", "Yolo", "Butte", "El Dorado", "Shasta"]
bucket_6 = ["Imperial", "Madera", "Kings", "Napa", "Humboldt", "Nevada"]
bucket_7 = ["Sutter", "Mendocino", "Yuba", "Lake", "Tehama", "San Benito"]
bucket_8 = ["Tuolumne", "Calaveras", "Siskiyou", "Amador", "Lassen", "Glenn"]
buckets = [bucket_1, bucket_2, bucket_3, bucket_4, bucket_5, bucket_6, bucket_7, bucket_8]
```

```python
df = import_csv("death_counts.csv")
df
```

```python
# Visualize deaths across all 10 buckets
for i in range(len(buckets)):
    graph_bucket(df, buckets[i])
    plt.title(f"OD Deaths 2018 - 2021 in Bucket {i + 1} Counties")
    plt.xlabel("Year")
    plt.ylabel("Deaths per year")
    plt.show()
```

```python
rate_df = import_csv("death_rates.csv")
rate_df
```

```python
# Visualize death rates across all 10 buckets
for i in range(len(buckets)):
    graph_bucket(rate_df, buckets[i])
    plt.title(f"OD Death Rate 2018 - 2021 in Bucket {i + 1} Counties")
    plt.xlabel("Year")
    plt.ylabel("Deaths per 100,000")
    plt.show()
```
