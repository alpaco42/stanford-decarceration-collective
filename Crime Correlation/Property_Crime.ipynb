# Notebook to get the property crime dataframe

```python
import pandas as pd
```

```python
df = pd.read_csv("county_crimes.csv", header=1)
df = df.iloc[:, 0:6] # this value is harcoded to get 2017 - 2021
df.head(10)
```

```python
property_crime = df[df["County / Year"] == "- All Property \nCrimes"]
property_crime.head()
```

```python
# jank way to hardcode the indices into the 
counties = pd.read_csv("death_rates.csv", header=1) # takes in previously used dataset with counties
counties = counties.iloc[:, 0] # grabs first column

assert(len(property_crime) == len(counties)) # for validation
for i in range(len(property_crime)):
    property_crime.iloc[i][0] = counties[i]
property_crime.columns = ["County" if "/" in x else x for x in property_crime.columns]
property_crime.head()
```
