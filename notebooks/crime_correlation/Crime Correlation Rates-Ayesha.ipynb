# Crime Correlation Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define 10 population buckets

pop_bucket_1 = ["Los Angeles", "San Diego", "Orange", "Riverside", "San Bernardino", "Santa Clara"]
pop_bucket_2 = ["Alameda", "Sacramento", "Contra Costa", "Fresno", "Kern", "San Francisco"]
pop_bucket_3 = ["Ventura", "San Joaquin", "San Mateo", "Stanislaus", "Sonoma", "Tulare"]
pop_bucket_4 = ["Solano", "Santa Barbara", "Monterey", "Placer", "San Luis Obispo", "Merced"]
pop_bucket_5 = ["Santa Cruz", "Marin", "Yolo", "Butte", "El Dorado", "Shasta"]
pop_bucket_6 = ["Imperial", "Madera", "Kings", "Napa", "Humboldt", "Nevada"]
pop_bucket_7 = ["Sutter", "Mendocino", "Yuba", "Lake", "Tehama", "San Benito"]
pop_bucket_8 = ["Tuolumne", "Calaveras", "Siskiyou", "Amador", "Lassen", "Glenn"]
pop_bucket_9 = ["Del Norte", "Colusa", "Plumas", "Inyo", "Mariposa", "Trinity"]
pop_bucket_10 = ["Mono", "Modoc", "Sierra", "Alpine"]
pop_buckets = [pop_bucket_1, pop_bucket_2, pop_bucket_3, pop_bucket_4, pop_bucket_5, 
               pop_bucket_6, pop_bucket_7, pop_bucket_8, pop_bucket_9, pop_bucket_10]
```

```python
# Define wealth buckets:

wealth_bucket_1 = ["Santa Clara", "Marin", "San Mateo", "Contra Costa", "San Francisco", "Ventura", "Orange", "Alameda", "Placer", "Napa"]
wealth_bucket_2 = ["El Dorado", "San Benito", "Solano", "Santa Cruz", "San Diego", "Sonoma", "Santa Barbara", "Mono", "Alpine", "San Luis Obispo"]
wealth_bucket_3 = ["Monterey", "Nevada", "Riverside", "Los Angeles", "Sacramento", "Yolo", "Calaveras", "San Bernardino", "Lassen", "San Joaquin"]
wealth_bucket_4 = ["Amador", "Sutter", "Mariposa", "Colusa", "Stanislaus", "Kern", "Tuolumne", "Plumas", "Kings", "Inyo"]
wealth_bucket_5 = ["Madera", "Yuba", "Fresno", "Shasta", "Mendocino", "Butte", "Sierra", "Merced", "Tulare", "Tehama"]
wealth_bucket_6 = ["Humboldt", "Imperial", "Glenn", "Del Norte", "Modoc", "Siskiyou", "Trinity", "Lake"]

wealth_buckets = [wealth_bucket_1, wealth_bucket_2, wealth_bucket_3, wealth_bucket_4, wealth_bucket_5, wealth_bucket_6]
```

```python
# Helper functions to graph buckets:

# graph a single county
def graph_county(df, name, years):
    county = df[df["County"] == name].iloc[0][1:]
    plt.plot(years, county, label=name)
    plt.legend(bbox_to_anchor=(1, 1))
    plt.xticks(years)
    
# graph a bucket of counties
def graph_bucket(df, bucket, years):
    for county in bucket: 
        graph_county(df, county, years)
        
# graph every bucket:
def graph_all_buckets(df, pop_buckets, title, y_axis_label):
    years = (df.columns)[1:]
    for i in range(len(pop_buckets)):
        graph_bucket(df, pop_buckets[i], years)
        plt.title(f"{title} in Bucket {i + 1} Counties")
        plt.xlabel("Year")
        plt.ylabel(f"{y_axis_label}")
        plt.show()
```

```python
# Grab the data from the CSV file, strip any excess whitespace, and plot it:

health_factors = pd.read_csv("health_factors.csv")
health_factors["County"] = health_factors["County"].str.strip()
graph_all_buckets(health_factors, pop_buckets, "Health Factors Ranks", "Ranks (0-58)")
```

```python
unemployment_rate = pd.read_csv("unemployment.csv")

unemployment_rate["County"] = unemployment_rate["County"].str.strip()
graph_all_buckets(unemployment_rate, pop_buckets, "Unemployment Rates", "Percent")
```
