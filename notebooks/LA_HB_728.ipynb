# Turning sentence_length into transitions_data

```python
sentence_length
```

```python
sentence_length.index
```

```python
long_finite_sentences = sentence_length.loc['50+'].astype(float).sum() / (
    sentence_length.loc[['50+', 'death', 'life']].astype(float).sum().sum()
)
long_finite_sentences
```

```python
transitions_data = sentence_length.copy()
transitions_data = transitions_data.iloc[1:]

for i in range(len(crime_types)):
    num = '' if i == 0 else '.' + str(i)
    transitions_data[crime_types[i]] = transitions_data['Black' + num].astype(float) + transitions_data['White' + num].astype(float)
    
transitions_data = transitions_data[crime_types]

transitions_data.loc['50-50'] = transitions_data.loc['50+'] + transitions_data.loc['death'] + transitions_data.loc['life']
transitions_data = transitions_data.drop(['death', 'life', '50+'])

transitions_data
```

```python
transitions_data = transitions_data.reset_index()
```

```python
transitions_data = transitions_data.rename(columns={'index': 'compartment_duration'})
```

```python
def punk_jr(compartment_duration):
    compartment_duration = compartment_duration.split('-')
    start = float(compartment_duration[0])
    final_duration = start
    
    if len(compartment_duration) > 1:
        end = float(compartment_duration[1])
        final_duration = -start + end + 1
        
    return final_duration * 12

transitions_data_jr = transitions_data.copy()
transitions_data_jr["months_in_group"] = transitions_data_jr['compartment_duration'].apply(punk_jr)


# disallow sentences less than 6 months
transitions_data_jr.loc[0, 'compartment_duration'] = '0.5-2'
transitions_data_jr.loc[0, 'months_in_group'] = 30.0

transitions_data_jr
```

```python
transitions_data_jr.property.sum() + transitions_data_jr.violent.sum() + transitions_data_jr.drug.sum() + transitions_data_jr.others.sum()
```

```python
transitions_jr_the_second = pd.DataFrame()
crime_types = ['violent', 'property', 'drug', 'others']
prev_months = 6

for i, row in transitions_data_jr.iterrows():
    for crime_type in crime_types:
        for month in range(int(prev_months), int(prev_months + row['months_in_group'])):
            new_row = {'compartment_duration': month, 'crime_type': crime_type, 'total_population': row[crime_type]/(month * row['months_in_group']),
                   'compartment': 'prison', 'outflow_to': 'release'}
            transitions_jr_the_second = transitions_jr_the_second.append(new_row, ignore_index=True)
    prev_months += row['months_in_group']

transitions_jr_the_second = transitions_jr_the_second.groupby(['crime_type', 'compartment_duration'], as_index=False).sum()
transitions_jr_the_second['compartment'] = 'prison' 
transitions_jr_the_second['outflow_to'] = 'release' 
```

```python
for crime_type in transitions_jr_the_second.crime_type.unique():
    crime_data = transitions_jr_the_second[transitions_jr_the_second.crime_type == crime_type]
    avg_LOS = (
    crime_data.compartment_duration * crime_data.total_population
    ).sum()/crime_data.total_population.sum()
    print(f'{crime_type} avg sent: {avg_LOS / 12}')
```

```python
transitions_life = transitions_jr_the_second[transitions_jr_the_second.compartment_duration > 50 * 12]


#breakdown of life sentences by crime type
lifer_by_crime_type = (transitions_life.groupby('crime_type').sum() / transitions_jr_the_second.groupby('crime_type').sum()).total_population

transitions_life.total_population *= 1 - long_finite_sentences
transitions_jr_the_second.loc[transitions_jr_the_second.compartment_duration > 50 * 12, 'total_population'] *= long_finite_sentences
transitions_life.loc[:,'crime_type'] = 'life'
transitions_jr_the_second = transitions_jr_the_second.append(transitions_life)
```

```python
#change admissions data to capture lifers separately
lifer_by_crime_type
outflows_data

before = outflows_data.total_population.sum()

for time_step in outflows_data.time_step.unique():
    lifer_population = 0
    for crime_type in crime_types:
        crime_type_lifers = lifer_by_crime_type[crime_type] * outflows_data[
            (outflows_data.time_step == time_step) & (outflows_data.crime_type == crime_type)
        ].total_population.sum()
        
        lifer_population += crime_type_lifers
        
        outflows_data.loc[
            (outflows_data.time_step == time_step) & (outflows_data.crime_type == crime_type), 'total_population'
        ] -= crime_type_lifers
    outflows_data = outflows_data.append({
        'crime_type': 'life',
        'time_step': time_step,
        'total_population': lifer_population,
        'compartment': 'pretrial',
        'outflow_to': 'prison'
    }, ignore_index=True)

assert(outflows_data.total_population.sum() - before < 0.0000001)
 
```

```python
# grant rate from here: https://s32082.pcdn.co/wp-content/uploads/2020/02/FY-2019-Annual-Report-Final.pdf
# The logic in here might be broken, i'm honestly not sure. The policy does actually do something, but it's 
# fishy that the percent to be reallocated thing being printed is 0. Investigate before re-using!
PAROLE_GRANT_RATE = 0.54

age_breakdown = pd.DataFrame({
    'age': ['16', '17', '18', '19-20', '21-24', '25+'],
    '51+ count': [3,7,18,72,222, 688]
})

policy_list = []

for crime_type in crime_types:
    policy_transitions = transitions_jr_the_second[transitions_jr_the_second.crime_type == crime_type]

    long_sentence_count = policy_transitions[policy_transitions.compartment_duration > 600].total_population.sum()

    transitions_to_reallocate = long_sentence_count * PAROLE_GRANT_RATE
    print(f'percent of transitions to reallocate: {transitions_to_reallocate / policy_transitions.total_population.sum() * 100}')

    policy_transitions = policy_transitions.append(pd.DataFrame({
        'compartment': ['prison'] * 6,
        'outflow_to': ['release'] * 6,
        'crime_type': [crime_type] * 6,
        'total_population': age_breakdown['51+ count'] / age_breakdown['51+ count'].sum() * transitions_to_reallocate,
        'compartment_duration': [29 * 12, 28 * 12, 27 * 12, 25.5 * 12, 22.5 * 12, 20 * 12]
    }))

    policy_transitions.loc[
        policy_transitions.compartment_duration > 600, 'total_population'
    ] *= 1 - long_finite_sentences * PAROLE_GRANT_RATE
    
    assert policy_transitions.total_population.sum() == transitions_jr_the_second[transitions_jr_the_second.crime_type == crime_type].total_population.sum()

    def grant_parole_to_long_sentences():
        return partial(TransitionTable.use_alternate_transitions_data, 
                       alternate_historical_transitions=policy_transitions,
                       retroactive=True)

    policy_list.append(SparkPolicy(policy_fn=grant_parole_to_long_sentences(),
                                           spark_compartment='prison',
                                           sub_population={'crime_type': crime_type},
                                           policy_ts=36, #jan 2023
                                           apply_retroactive=True))


test_results = spark_sim.simulate_policy(policy_list, 'prison')
```
